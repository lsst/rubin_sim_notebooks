```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import glob
from rubin_sim.data import get_data_dir
from rubin_sim.movingObjects import readOrbits
```

```python
dd = get_data_dir()
ss_files = glob.glob(dd+'/orbits/*.txt')
```

```python
for filename in ss_files:
    fig, ax = plt.subplots()
    orbits = readOrbits(filename)
    try:
        sc = ax.scatter(orbits.orbits['a'], orbits.orbits['e'], c=orbits.orbits['inc'], alpha=0.2)
        ax.set_xlabel('a (AU)')
    except:
        sc = ax.scatter(orbits.orbits['q'], orbits.orbits['e'], c=orbits.orbits['inc'], alpha=0.2)
        ax.set_xlabel('q (AU)')
    
    ax.set_ylabel('e')
    cb = fig.colorbar(sc)
    cb.set_label('Inc (deg)')
    ax.set_title(filename.split('/')[-1].replace('.txt',''))
```
