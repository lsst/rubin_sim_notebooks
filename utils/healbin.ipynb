```python
from rubin_scheduler.utils import healbin
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import healpy as hp
```

```python
nside = 32

ra = np.arange(0,360, 1)
dec = np.arange(-90,90,0.5)

vals =np.arange(360) + 10

```

```python
hp_map = healbin(ra, dec, vals, nside, reduce_func=np.mean)
```

```python
hp.mollview(hp_map, unit='mean val', title='title')
```

```python
hp_map = healbin(ra, dec, vals, nside, reduce_func=np.sum)
```

```python
hp.mollview(hp_map, unit='sum val', title='title')
```
